```python
#使用selenium webdriver调用chrome, 访问www.baidu.com，
#并用程序控制在检索框输入自己的姓名，抽取搜索结果列表的内容和分页的链接列表。
from selenium import webdriver
from pyquery import PyQuery as pq

dr = webdriver.Chrome()
#get方法来打开指定网址
dr.get("http://www.baidu.com/")
#id选择到搜索框
keyword=dr.find_element_by_id('kw')
#搜多框键入名字
keyword.send_keys('王宁')
#点击搜索
button=dr.find_element_by_id('su')
button.click()
#搜索已完成

#左边搜索列表的内容们
dr.find_element_by_id('content_left')
```

```python
dir(dr)
```

```python
s=dr.page_source
print(s)
```

```python
doc=pq(s)
lst=doc('#content_left .c-container h3 a em')
print(lst)
```

```python
listw=doc('#content_left .c-container h3 a').items()
for li in listw:
    print(li)
```

```python
dr.find_element_by_id('page')
```

```python
#爬取分页
f=dr.page_source
doc2=pq(f)
list2=doc2('#page a').items()
for li in list2:
    print(li)
```
