### 计算文章中的英文单词数量-真的token有多少

```python
print('文本总长度：',len(text1))
print('不重复token长度：',len(set(text1)))
print('不区分大小写token长度：',len(set([word.lower() for word in text1])))
print('纯单词token长度：',len(set([word.lower() for word in text1 if word.isalpha()])))
```
