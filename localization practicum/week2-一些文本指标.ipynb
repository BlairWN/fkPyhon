### 由于高频和超低频单词都没法反应文信息，所以找寻别的单词

```python
#长度>15的单词
V=set(text1)
longWords=[w for w in V if len(w)>15]
sorted(longWords)
```

```python
#查询文本里超过15字母的单词
def longwords(text):
    V=set(text)
    longWord=[w for w in V if len(w)>15]
    return sorted(longWord)
```

```python
#查询词频>7，频数>7的单词，为了过滤掉那些the啥的
def infwords(text):
    fdistset=FreqDist(text)
    infw=sorted([w for w in set(text) if len(w)>7 and fdistset[w]>7])
    return infw
```
